```python
#5. What are iterators in Python and how do they differ from iterables?
#In Python, iterables are objects that can be iterated over, meaning they can be used in a for loop. Examples include lists, tuples, dictionaries, and strings.

#Iterators are objects that implement the __iter__() and __next__() methods. When an iterator is created from an iterable using the iter() function, it allows you to iterate over the elements one by one, maintaining the state of iteration.

#Differences:

#Iterables:

#Can be iterated over.
#Do not have __next__() method.
#Examples: lists, tuples, dictionaries, strings.
#Iterators:

#Created from iterables using iter() function.
#Have __iter__() and __next__() methods.
#Keep track of the current state of iteration.
Example of using an iterator:
my_list = [1, 2, 3]
my_iter = iter(my_list)

print(next(my_iter))  # Output: 1
print(next(my_iter))  # Output: 2
print(next(my_iter))  # Output: 3
```

```python
#6. Explain the concept of generators in Python and how they are defined.
#In Python, generators are a convenient way to create iterators. They are defined using a function that contains one or more yield statements. When the function is called, it returns a generator object without executing the function body. Each time the next() function is called on the generator object, the function runs until it encounters a yield statement, which pauses the function and saves its state. The generator remembers this state and continues from where it left off when next() is called again.

#Generators are memory efficient compared to lists because they generate values on the fly instead of storing them in memory. They are particularly useful for generating a sequence of values over time.

#Example of defining a generator using a function with yield:

def my_generator():
    yield 1
    yield 2
    yield 3

gen = my_generator()
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
print(next(gen))  # Output: 3
```

```python
#7. What are the advantages of using generators over regular functions?
#The advantages of using generators over regular functions in Python include:

#Memory Efficiency:

#Generators produce values on-the-fly and do not store all values in memory at once.
#Useful for generating large sequences of values without occupying a lot of memory.
#Lazy Evaluation:

#Values are generated only when needed, making generators suitable for scenarios where values are consumed one at a time.
#Can be more efficient in terms of performance.
#Simplified Code:

#Generators are concise and often require less code compared to using regular functions.
#Iterative Processing:

#Generators are well-suited for iterative processing, such as handling large datasets or infinite sequences.
#State Retention:

#Generators automatically retain their state between function calls, eliminating the need for manual state management.
#Versatility:

#Generators can be used in a variety of contexts, including loops, comprehensions, and passing as arguments to other functions.
#Overall, generators offer a more efficient and elegant way to work with sequences of data in Python, especially when dealing with large or potentially infinite datasets.
```

```python
#9. Explain the purpose and usage of the `map()` function in Python.
#The map() function in Python is used to apply a function to each item in an iterable (such as a list) and return a new iterator that yields the results. It takes two arguments: the function to apply and the iterable to apply it to. The map() function returns a map object, which is an iterator.

#Example:
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
squared_numbers = map(square, numbers)

print(list(squared_numbers))  # Output: [1, 4, 9, 16]

#The map() function can also be used with lambda functions for more concise code:

#Example using lambda function:
numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x: x * x, numbers)

print(list(squared_numbers))  # Output: [1, 4, 9, 16]

#The map() function is useful for applying a specific operation to all elements of an iterable without needing to write explicit loops. It provides a functional programming approach to processing data in Python.



```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.
#code
def is_prime(number):
    if number < 2:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

# Check numbers from 1 to 200 for primality
for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is prime")
    else:
        print(f"{num} is not prime")
```

```python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
#code:
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        else:
            self.count += 1
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            return result

# Example usage
fibonacci = FibonacciIterator(10)
for num in fibonacci:
    print(num)
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
#code:
def powers_of_two(exponent):
    return (2 ** i for i in range(exponent + 1))

# Example usage:
for power in powers_of_two(5):
    print(power)
```

```python
#9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
#code:
 # List of temperatures in Celsius
celsius_temperatures = [0, 10, 20, 30, 40]

# Conversion function from Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# Use map() to convert the list of temperatures from Celsius to Fahrenheit
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

print(fahrenheit_temperatures)
```

```python
#10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.
#code:
# Given string
input_string = "hello world"

# Function to filter out vowels
def remove_vowels(char):
    vowels = 'aeiouAEIOU'
    return char not in vowels

# Use filter() to remove vowels from the string
result = ''.join(filter(remove_vowels, input_string))

print(result)
```
